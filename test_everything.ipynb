```python
import base64
import os

def image_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    return encoded_string

image_path = os.path.join("images","4_.jpeg")
encoded_image = image_to_base64(image_path)

# Hin th chui base64
print(encoded_image)
```

```python
import requests
```

```python
def split_base64_string(base64_string, chunk_size=256):
    return [base64_string[i:i + chunk_size] for i in range(0, len(base64_string), chunk_size)]
```

```python
import requests

# Gi s split_base64_string l hm chia nh chui base64
for part in split_base64_string(encoded_image):
    print(part)
    response = requests.post(
        "http://103.179.191.94:8000/upload_part/",
        json={"part": part}  # m bo gi JSON ng cch
    )
    print("-" * 1000)
    print(response.json())
```

```python
response = requests.get("http://103.179.191.94:8000/complete_upload/")
print(response.json())
```
